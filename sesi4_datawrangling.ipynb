Descriptive Analysis

- Data wisatawan laki - laki dalam bentuk Bar Chart (Diagram Batang)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_excel("data_mancanegara.xlsx", sheet_name="Gender Age")

df.dropna(axis=0, inplace=True)

data_array = df.to_numpy()
transposed_array = data_array.transpose()
sorted_array = transposed_array[:, np.argsort( transposed_array[2] )]

x = np.array(sorted_array[2])
y = np.array(sorted_array[0])

plt.bar(x, y, color='blue') 
plt.xlabel("Laki laki") 
plt.ylabel("Data Wisatawan Mancanegara menurut Jenis Kelamin") 
plt.show()
```

- Data wisatawan perempuan dalam bentuk Bar Chart (Diagram Batang)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df = pd.read_excel("data_mancanegara.xlsx", sheet_name="Gender Age")

df.dropna(axis=0, inplace=True)

data_array = df.to_numpy()
transposed_array = data_array.transpose()
sorted_array = transposed_array[:, np.argsort( transposed_array[1] )]


x = np.array(sorted_array[1])
y = np.array(sorted_array[0])

plt.bar(x, y, color='red') 
plt.xlabel("Perempuan") 
plt.ylabel("Data Wisatawan Mancanegara menurut Jenis Kelamin") 
plt.show()
```

- Mean (Rata- rata)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Membaca data dari Excel
df = pd.read_excel("data_mancanegara.xlsx", sheet_name="Gender Age")

# Menghitung mean untuk kolom "Perempuan" dan "Laki-Laki"
mean_perempuan = df["Perempuan"].mean()
mean_laki_laki = df["Laki-Laki"].mean()

# Menampilkan hasil
print(f"Mean Perempuan: {mean_perempuan}")
print(f"Mean Laki-Laki: {mean_laki_laki}")
```

Median

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Membaca data dari Excel
df = pd.read_excel("data_mancanegara.xlsx", sheet_name="Gender Age")

# Menghitung median untuk kolom "Perempuan" dan "Laki-Laki"
median_perempuan = df["Perempuan"].median()
median_laki_laki = df["Laki-Laki"].median()

# Menampilkan hasil
print(f"Median Perempuan: {median_perempuan}")
print(f"Median Laki-Laki: {median_laki_laki}")
```

Mode (Modus)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Membaca data dari Excel
df = pd.read_excel("data_mancanegara.xlsx", sheet_name="Gender Age")

# Menghitung modus untuk kolom "Perempuan" dan "Laki-Laki"
modus_perempuan = df["Perempuan"].mode()[0]  # Ambil nilai modus pertama
modus_laki_laki = df["Laki-Laki"].mode()[0]  # Ambil nilai modus pertama

# Menampilkan hasil
print(f"Modus Perempuan: {modus_perempuan}")
print(f"Modus Laki-Laki: {modus_laki_laki}")
```
